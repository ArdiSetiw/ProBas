## a. Buat Tabel Distribusi Frekuensi dan Histogram Variabe reading_score dan writing_score dengan jumlah dan lebar interval sebesar 10.

### Table Distribusi Frekuensi

```python
data_reading = df['reading score'].value_counts(sort=False,bins=np.arange(0,110,10)).rename_axis('reading score').reset_index(name='frequency')

data_writing = df["writing score"].value_counts(sort=False,bins=np.arange(0, 110, 10)).rename_axis("writing score").reset_index(name="frequency")

data = {'reading score':data_reading['reading score'],'reading freq':data_reading['frequency'],
        'writing score':data_writing['writing score'],'writing freq':data_writing['frequency']}

table_freq = pd.DataFrame(data)
table_freq
```

### Histogram Reading Score dan Writing Score

```python
fig, axes = plt.subplots(1, 2, sharex=True, figsize=(15,5))

sns.histplot(
    ax = axes[0],
    data = df,
    x = 'reading score',
    bins = np.arange(0,110,10),
    kde=True,color='green'
)

sns.histplot(
    ax = axes[1],
    data = df,
    x = 'writing score',
    bins = np.arange(0,110,10),
    kde=True
)


axes[0].set_title('Reading Score Histogram')
axes[0].set_ylabel('Frequency')
axes[1].set_title('Writing Score Histogram')
axes[1].set_ylabel('Frequency')
```

## b. Tentukan nilai Tendensi Central: Rata-rata, Median, Modus, Min dan Max data kedua variable tersebut.

```python
# Tendensi Central Reading Score
mean_read = df['reading score'].mean()
median_read = df['reading score'].median()
modus_read = df['reading score'].mode().values[0]
min_read  = df['reading score'].min()
max_read = df['reading score'].max()

# Tendensi Central Writing Score
mean_write = df['writing score'].mean()
median_write = df['writing score'].median()
modus_write = df['writing score'].mode().values[0]
min_write = df['writing score'].min()
max_write = df['writing score'].max()


tendensi_central = pd.DataFrame({'reading score':[mean_read,median_read,modus_read,min_read,max_read],
                                'writing score':[mean_write,median_write,modus_write,min_write,max_write]
                                },index = ['rata-rata','median','modus','min','max'])

tendensi_central
```

## c. Tentukan nilai Variasi: Jangkauan Range, Kisaran Interkuartil, Varians, Standar Deviasi data reading_score dan writing_score.

```python
# Nilai Variasi Reading Score
range_read = df['reading score'].max() - df['reading score'].min()
iqr_read = df['reading score'].quantile(0.75) - df['reading score'].quantile(0.25)
var_read = df['reading score'].var()
std_read = df['reading score'].std()

# Nilai Variasi Writing Score
range_write = df['writing score'].max() - df['writing score'].min()
iqr_write = df['writing score'].quantile(0.75) - df['writing score'].quantile(0.25)
var_write = df['writing score'].var()
std_write = df['writing score'].std()


nilai_variasi = pd.DataFrame({'reading score':[range_read,iqr_read,var_read,std_read],
                              'writing score':[range_write,iqr_write,var_write,std_write]
                             },index = ['Jangkauan Range','Jangkauan Quartile','Variasi','Standart Deviasi'])

nilai_variasi
```

## d. Buat Grafik data Variabel reading_score dan writing_score: Histogram, BOX PLOT, dan Tentukan Nilai Skewness dan Kurtosis. Buat interpretasi saudara tentang perbedaan distribusi kedua variable tersebut, terutama pada gambar BOX PLOT dan nilai Skewness - Kurtosisnya.

## Histogram Reading Score dan Writing Score

```python
fig, axes = plt.subplots(1, 2, sharex=True, figsize=(15,5))

sns.histplot(ax=axes[0],data=df,x='reading score',color='green',kde=True,bins = np.arange(0,110,10))
sns.histplot(ax=axes[1],data=df,x='writing score',kde=True,bins = np.arange(0,110,10),)

axes[0].set_title('Reading Score Histogram')
axes[0].set_ylabel('Frequency')
axes[1].set_title('Writing Score Histogram')
axes[1].set_ylabel('Frequency')
```

### Boxplot Reading Score dan writing Score

```python
sns.boxplot(data=df[["reading score", "writing score"]], orient="v")

plt.title('Boxplot Reading and Writing Score')

print("----reading score----")
print("Nilai Skewness  :",df['reading score'].skew())
print("Nilai Kurtosis  :",df['reading score'].kurt())
print("")
print("----writing score----")
print("Nilai Skewness  :",df['writing score'].skew())
print("Nilai Kurtosis  :",df['writing score'].kurt())
```
